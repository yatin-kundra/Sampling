```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTETomek
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Creditcard_data.csv')
y = df['Class']
X = df.drop('Class',axis =1)
del df
```

```python
X
```

## Random Oversampling

```python
ros = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_ros, y_ros = ros.fit_resample(X, y)
display(X_ros)
```

## Random Undersampling

```python
rus = RandomUnderSampler(sampling_strategy='auto', random_state=42)
X_rus, y_rus = rus.fit_resample(X, y)
display(X_rus)
```

## SMOTE

```python
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)
display(X_smote)
```

## Stratified Shuffle Sampling

```python
stratified_sampler = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
for train_index, _ in stratified_sampler.split(X, y):
    X_stratified, y_stratified = X.iloc[train_index], y.iloc[train_index]
display(X_stratified)
```

## SMOTETomek

```python
smote_tomek = SMOTETomek(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote_tomek.fit_resample(X, y)
display(X_resampled)
```

## Training Models

```python
model_names = ['RandomForestClassifier', 'LogisticRegression', 'SVC','KNeighborsClassifier','DecisionTreeClassifier']
row_names = ['Random Oversampled', 'Random Undersampled', 'SMOTE sampled','Stratified sampled','SMOTETomek']
metrics = pd.DataFrame(0.0, index=row_names, columns=model_names)
metrics
```

```python
random_seed =42
models = [
    RandomForestClassifier(n_estimators=100, max_depth=2, random_state=random_seed),
    LogisticRegression(solver='liblinear', random_state=random_seed),
    SVC(kernel='linear', C=1.0, random_state=random_seed),
    KNeighborsClassifier(n_neighbors=5, algorithm='auto'),
    DecisionTreeClassifier(max_depth=2, random_state=random_seed)
]
model_zip = zip(models,['RandomForestClassifier', 'LogisticRegression', 'SVC','KNeighborsClassifier','DecisionTreeClassifier'])
metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ros,y_ros, test_size=0.2, random_state=42)
for j, (model,model_name) in enumerate(model_zip):
    model.fit(X_train, y_train)
    y_pred_sample = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred_sample)
    print(accuracy)
    metrics.loc['Random Oversampled',model_name] = accuracy
metrics
```

```python
model_zip = zip(models,['RandomForestClassifier', 'LogisticRegression', 'SVC','KNeighborsClassifier','DecisionTreeClassifier'])
X_train, X_test, y_train, y_test = train_test_split(X_rus,y_rus, test_size=0.2, random_state=42)
for j, (model,model_name) in enumerate(model_zip):
    model.fit(X_train, y_train)
    y_pred_sample = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred_sample)
    metrics.loc['Random Undersampled',model_name] = accuracy
metrics
```

```python
model_zip = zip(models,['RandomForestClassifier', 'LogisticRegression', 'SVC','KNeighborsClassifier','DecisionTreeClassifier'])
X_train, X_test, y_train, y_test = train_test_split(X_smote,y_smote, test_size=0.2, random_state=42)
for j, (model,model_name) in enumerate(model_zip):
    model.fit(X_train, y_train)
    y_pred_sample = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred_sample)
    print(accuracy)
    metrics.loc['SMOTE sampled',model_name] = accuracy
metrics
```

```python
model_zip = zip(models,['RandomForestClassifier', 'LogisticRegression', 'SVC','KNeighborsClassifier','DecisionTreeClassifier'])
X_train, X_test, y_train, y_test = train_test_split(X_stratified,y_stratified, test_size=0.2, random_state=42)
for j, (model,model_name) in enumerate(model_zip):
    model.fit(X_train, y_train)
    y_pred_sample = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred_sample)
    print(accuracy)
    metrics.loc['Stratified sampled',model_name] = accuracy
        
```

```python
model_zip = zip(models,['RandomForestClassifier', 'LogisticRegression', 'SVC','KNeighborsClassifier','DecisionTreeClassifier'])
X_train, X_test, y_train, y_test = train_test_split(X_resampled,y_resampled, test_size=0.2, random_state=42)
for j, (model,model_name) in enumerate(model_zip):
    model.fit(X_train, y_train)
    y_pred_sample = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred_sample)
    print(accuracy)
    metrics.loc['SMOTETomek',model_name] = accuracy
metrics
```
